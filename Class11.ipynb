# **Sentiment Analysis**

```python
# Importing all the necessary resources:
! pip3 install textblob
! pip3 install datasets

from datasets import load_dataset
from textblob import TextBlob
```

```python
# Loading the dataset:
dataset = load_dataset('Censius-AI/ECommerce-Women-Clothing-Reviews')
dataset

reviews = dataset['train']['Review Text']
reviews[0]

stars = dataset['train']['Rating']
stars[0]

len(reviews)
len(stars)

TextBlob(reviews[0]).sentiment
```

```python
for i, review in enumerate(reviews):
  if review:
    polarity = TextBlob(review).sentiment.polarity
    print(stars[i], polarity)
```

```python
sentiments = {'positive': [5, 4], 'neutral': [3], 'negative': [1, 2]}
strange_reviews = []

for i, review in enumerate(reviews):
  if review:
    polarity = TextBlob(review).sentiment.polarity
    if polarity > 0.2:
      sentiment = 'positive'
    elif -0.2 <= polarity <= 0.2:
      sentiment = 'neutral'
    elif polarity < -0.2:
      sentiment = 'negative'
      review_stars = stars[i]
      if review_stars not in sentiments[sentiment]:
        strange_reviews.append((stars[i], polarity))
```

```python
print(strange_reviews)
len(strange_reviews)
len(strange_reviews) / len(reviews)
```
