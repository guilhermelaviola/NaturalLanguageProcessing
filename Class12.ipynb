# **Machine Learning and Deep Learning for Natural Language**

```python
# Importing all the necessary resources:
! pip3 install transformers
! pip3 install datasets


from datasets import load_dataset
from transformers import pipeline
```

```python
# Loading the dataset:
dataset = load_dataset("Censius-AI/ECommerce-Women-Clothing-Reviews")
dataset

reviews = dataset['train']['Review Text']
reviews[0]

stars = dataset['train']['Rating']
stars[0]

len(reviews)
len(stars)
```

```python
classificator = pipeline(
    model='lxyuan/distilbert-base-multilingual-cased-sentiments-student',
    return_all_scores=True,
)

classificator('Achei o filme excelente')
classificator('Achei o filme horrível!')
classificator('Achei o filme ok')
classificator('Não sei dizer se gostei do filme')
classificator(['Não sei dizer se gostei do filme', 'Adorei o filme'])
print([x for x in reviews if not x])
reviews = [x if x else '' for x in reviews]
print([x for x in reviews if not x])
```

```python
# For all the 23.486 reviews, the classification would take 1 hour:
classifications = classificator(reviews[:100])

for i, review in enumerate(reviews[:100]):
  if review:
    print((stars[i], classifications[i]))
```
