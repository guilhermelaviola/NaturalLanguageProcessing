```python
# Installing and loading datasets package:
! pip3 install datasets
from datasets import load_dataset
```

```python
# Academic dataset:
academic_dataset = load_dataset('capes')

academic_dataset

academic_dataset['train']['translation']

academic_dataset = [text['en'] for text in academic_dataset['train']['translation']]
```

```python
# E-Commerce dataset:
ecommerce_dataset = load_dataset('ruanchaves/b2w-reviews01')

ecommerce_dataset

ecommerce_dataset['train']['review_text']

ecommerce_dataset = [text for text in ecommerce_dataset['train']['review_text'] if isinstance(text, str)]
```

```python
# Text preprocessing:
academic_dataset

documents = {}

documents['academico'] = ' '.join(academic_dataset)

documents['academico'][:5000]

documents['ecommerce'] = ' '.join(ecommerce_dataset)

documents['ecommerce'][:5000]
```

```python
def pre_process(text):
  text = text.replace(",", "")
  text = text.replace(".", "")
  text = text.replace("(", "")
  text = text.replace(")", "")
  text = text.replace("!", "")
  text = text.replace("?", "")
  text = text.lower()
  text = text.split(" ")
  return text

documents['academico'] = pre_process(documents['academico'])
documents['ecommerce'] = pre_process(documents['ecommerce'])

documents['academico'][:20]

documents['ecommerce'][:20]
```
