# **Analysis and interpretation of Natural Language**

```python
# Opening and displaying the subtitle file:
with open('Hatari.English-WWW.MY-SUBS.CO.srt') as f:
  sub = f.read()

print(sub)
```

```python
# Transforming the subtitles in text only:
sentences = sub.split('\n\n')

sentences

sentences[0].split('\n', 2)

sentences[0].split('\n', 2)[2]
```

```python
text = ''

for sentence in sentences:
    text = text + ' ' + sentence.split('\n', 2)[2]
```

```python
# Cleaning the subtitle:
text = text.replace('\n', ' ')
text = text.replace('', '')
text = text.replace('', '')
text = text.replace('', '')
text = text.replace('', '')
```

```python
# Segmenting sentences:
sentences = text.split('. ')

sentences

sentences = [sentence + '.' for sentence in sentences]

sentences
```

```python
# Segmenting tokens:
# List comprehension:loop within a list that generates another list:
sentences = [sentence\
             .replace('.', ' . ')\
             .replace(',', ' , ')\
             .replace('!', ' ! ')\
             .replace('?', ' ? ')\
             .replace(':', ' : ')\
             .replace('  ', ' ')\
             .strip() for sentence in sentences]

sentences

tokenized_sentences = [sentence.split(' ') for sentence in sentences]

tokenized_sentences

tokenized_sentences[0]
```

```python
# Generating a list of keywords with YAKE:
! pip3 install yake
import yake

text

kw_extractor = yake.KeywordExtractor(lan='en', n=1)
keywords = kw_extractor.extract_keywords(text)

keywords

sorted(keywords, key=lambda x: x[1], reverse=True)
```

```python
# Generating a list of keywords with SPACY:
! pip3 install spacy
! python3 -m spacy download en_core_web_sm # Download the model

import spacy

# Verify the model location (optional, for debugging)
# print(spacy.info()['location'])

nlp = spacy.load('en_core_web_sm') # Loading the model
doc = nlp(text)

for token in set(doc.ents):
  print(token)
```
